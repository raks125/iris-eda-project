```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

iris=sns.load_dataset("iris")
print(iris)

print("first 5 records\n")
print(iris.head(),'\n')
print("description about dataset\n")
print(iris.describe(),'\n')
print("info about dataset\n")
print(iris.info(),'\n')

count=iris['sepal_length'].value_counts()
count1=iris['sepal_width'].value_counts()
count2=iris['petal_length'].value_counts()
count3=iris['petal_width'].value_counts()
count4=iris['species'].value_counts()
print("number of species\n",count4)
plt.pie(count4,labels=count4.index,autopct='%1.1f%%')
plt.title('species')
plt.show()

plt.hist(iris['sepal_length'])
plt.title('sepal_length')
plt.show()

plt.hist(iris['sepal_width'])
plt.title('sepal_width')
plt.show()

plt.hist(iris['petal_length'])
plt.title('petal_length')
plt.show()

plt.hist(iris['petal_width'])
plt.title('petal_width')
plt.show()

plt.scatter(iris['sepal_length'],iris['sepal_width'])
plt.title('sepal_length vs sepal_width')
plt.show()

plt.scatter(iris['petal_length'],iris['petal_width'])
plt.title('petal_length vs petal_width')
plt.show()
```
